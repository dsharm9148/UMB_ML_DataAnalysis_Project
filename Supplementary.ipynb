### Additional Evaluation Metrics Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the metrics you want to plot
metrics = ['f1-score', 'precision', 'recall']

# Loop through each metric
for metric in metrics:
    # Filter for the current metric
    metric_df = results_df[results_df['MetricName'] == metric]

    # Create a new column for the class pair
    metric_df['ClassPair'] = metric_df['Class1'] + ' vs ' + metric_df['Class2']

    # Sort the DataFrame by the metric values in descending order
    metric_df_sorted = metric_df.sort_values(by='Value', ascending=False)

    # Plot the metric as a bar graph
    plt.figure(figsize=(12, 8))
    sns.barplot(
        data=metric_df_sorted,
        x='ClassPair',
        y='Value',
        hue='Classifier',
        palette='Set2',
        errorbar=None
    )
    plt.title(f'{metric.capitalize()} by Class Pair and Classifier')
    plt.xlabel('Class Pair')
    plt.ylabel(metric.capitalize())
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.legend(title='Classifier')
    plt.show()
```
